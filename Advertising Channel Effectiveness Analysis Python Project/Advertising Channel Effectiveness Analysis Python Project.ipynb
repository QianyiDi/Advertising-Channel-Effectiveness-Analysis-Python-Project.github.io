```python
import matplotlib.pyplot as plt  
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.metrics import silhouette_score  
from sklearn.cluster import KMeans  
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder  
```

```python
raw_data = pd.read_excel("D:/从c盘转移的桌面/github/Advertising Channel Effectiveness Analysis Python Project/ad_performance.xlsx")
```

```python
raw_data2=raw_data.drop(['Average Dwell Time'],axis=1)
```

```python
cols =['Creative Material Type','Advertising Type', 'Cooperation Mode', 'Ad Size', 'Advertising Selling Point']
```

```python
model_ohe=OneHotEncoder(sparse=False)
ohe_matrix=model_ohe.fit_transform(raw_data2[cols])
```

```python
scale_matrix=raw_data2.iloc[:,1:7]
```

```python
model_scaler=MinMaxScaler()
data_scaled=model_scaler.fit_transform(scale_matrix)
```

```python
X=np.hstack((data_scaled,ohe_matrix))
score_list=list()
silhouette_int=-1
```

```python
for n_clusters in range(2,5):
    model_kmeans=KMeans(n_clusters=n_clusters)
    labels_tmp=model_kmeans.fit_predict(X)
    silhouette_tmp=silhouette_score(X,labels_tmp)
    if silhouette_tmp>silhouette_int:
        best_k=n_clusters #keep the best K
        print(best_k)
        silhouette_int=silhouette_tmp#always keep best parameters
        print(silhouette_int)
        best_kmeans=model_kmeans#keep our model
        print(best_kmeans)
        cluster_labels_k=labels_tmp#keep the best model label
        print(cluster_labels_k)
    score_list.append([n_clusters,silhouette_tmp])
np.array(score_list)
```

```python
raw_data2['cluster']=cluster_labels_k
ratio=raw_data2.groupby('cluster')['Channel Code'].count()/raw_data2.shape[0]
```

```python
cluster_features=[]
for i in range(best_k):
    label_data=raw_data2[raw_data2['cluster']==i]#every group data
    part1_data=label_data.iloc[:,1:7]
    print(part1_data)
    part1_desc=part1_data.describe().round(3)
    print(part1_desc)
    merge_data1=part1_desc.iloc[2,:]
    print(merge_data1)
    part2_data = label_data.iloc[:, 7:-1]  # get string type data
    part2_desc = part2_data.describe(include='all') 
    merge_data2 = part2_desc.iloc[2, :]
    print(merge_data2)
    result=pd.concat([merge_data1,merge_data2],axis=0)
    print(result)
    cluster_features.append(result)
all_result=pd.DataFrame(cluster_features)
all_result['ratio']=ratio
print(all_result)
```
